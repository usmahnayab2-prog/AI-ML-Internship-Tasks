```python

# Task 01: Exploring and Visualizing the Iris DataSet
# Goal: To understand the dataset by looking at data and drawing graphs

# Import Tools
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Iris Datasets
df = sns.load_dataset("iris")
# check size of the data
df.shape
# show columns
df.columns
# look at first 5 rows
df.head()
# Get data information
df.info()
# get simple math summary
df.describe()
# now we draw garph Scatter Plot ( compare two things )
sns.scatterplot(
    x = "sepal_length",
    y = "petal_length",
    data = df,
    hue = "species",
)
plt.show()
# Histogram(How Often Values Appears)
df["sepal_length"].hist()
plt.title("sepal length")
plt.show()

# Box Plot (Find Strange Values)
sns.boxplot(x="species", y="petal_length", data=df)
plt.show()
```
